## Ejercicio 2 — Simular la trayectoria y estimar la distribución estacionaria (6 pts)

**Relación con RA1:** aquí conectas la teoría (Semana 1: proceso estocástico como familia de variables aleatorias) con la práctica computacional que ya dominas desde Simulación (generación de trayectorias aleatorias). La distribución estacionaria te dice, en el largo plazo, qué fracción del tiempo la máquina pasa en cada estado — información directamente accionable para el área de mantenimiento.

Simula **10,000 turnos** empezando en el Estado 0 (Operativa), y estima la distribución estacionaria empírica (la fracción de turnos que la máquina pasó en cada estado).

```python
# YOUR CODE HERE
def simulacion(n_pasos, t_matrix, estado_inicio, seed):
    rng_local = np.random.default_rng(seed)
    n_estados = t_matrix.shape[0]
    trayectoria = np.empty(n_pasos, dtype=int)
    trayectoria[0] = estado_inicio

    for i in range(1, n_pasos):
        estado_actual = trayectoria[i-1]
        trayectoria[i] = rng_local.choice(n_estados, p=t_matrix[estado_actual])

    return trayectoria

trayectoria = simulacion(10000, P, 0, np.random.randint(1, 100) )

distribucion_empirica = np.bincount(trayectoria, minlength=P.shape[0]) / len(trayectoria)

print(distribucion_empirica)
```

```python
### TEST
assert len(trayectoria) == 10000, "La trayectoria debe tener 10000 pasos"
assert trayectoria[0] == 0, "Debe iniciar en el Estado 0 (Operativa)"
assert set(trayectoria).issubset({0, 1, 2}), "Solo deben aparecer los estados 0, 1, 2"
# La distribución estacionaria analítica de esta cadena es aprox. [0.82, 0.13, 0.05]
assert abs(distribucion_empirica[0] - 0.82) < 0.03, "La fracción del Estado 0 se aleja demasiado del valor esperado"
print("Ejercicio 2: todas las verificaciones pasaron.")
```
